```python
import numpy as np
import turtle              
```

```python
def sierpinski_triangle_recursive(depth, length):
    if depth > 0:
        return sierpinski_triangle_recursive(depth-1, length/2) + sierpinski_triangle_recursive(depth-1, length/2) + sierpinski_triangle_recursive(depth-1, length/2)
    else:
        return np.sqrt(length**2-(length/2)**2)*length/2
```

```python
print(sierpinski_triangle_recursive(10, 1.0))
```

```python
# This is just for fun: Drawing Sierpinski triangles using turtle graphics
def draw_sierpinski(depth,length):
    if depth> 0:
        draw_sierpinski(depth-1,length/2)
        t.fd(length/2)
        draw_sierpinski(depth-1,length/2)
        t.bk(length/2)
        t.left(60)
        t.fd(length/2)
        t.right(60)
        draw_sierpinski(depth-1,length/2)
        t.left(60)
        t.bk(length/2)
        t.right(60)
    else:
        for i in range(0,3):
            t.fd(length)
            t.left(120)
```

```python
window = turtle.Screen()
t = turtle.Turtle()
t.speed(10)
draw_sierpinski(4,100)
turtle.done()
try:
    turtle.bye()
except:
    print("bye")
```
